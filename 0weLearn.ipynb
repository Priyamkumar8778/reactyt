#  Q. what is diff b/w library and framework ?
library

1. GSAP
2. Lenis
3. ReactJS

Framework

1. Nextjs
2. Angular
3. Remix


A **library** is a collection of pre-written functions or tools that you call whenever you need them, meaning **you control the flow of the program** and decide when to use the library's features. A **framework**, on the other hand, provides a predefined structure and controls the application's flow, meaning **the framework calls your code when needed**—a concept known as *Inversion of Control*. For example, **React** is primarily a library because you decide how and when to use its components, while frameworks like Angular or Next.js provide a complete application structure, routing, conventions, and lifecycle management that your code must follow.

# Q. RealDOM vs VirtualDOM
| Feature      | Real DOM                                               | Virtual DOM                                       |
| ------------ | ------------------------------------------------------ | ------------------------------------------------- |
| Definition   | Actual DOM tree maintained by the browser.             | Lightweight JavaScript representation of the DOM. |
| Update Speed | Slower because changes directly affect the browser UI. | Faster because changes are first made in memory.  |
| Re-rendering | Often updates large parts of the DOM.                  | Updates only the changed elements.                |
| Performance  | Can become slow for complex applications.              | More efficient for frequent UI updates.           |
| Memory Usage | Lower (only actual DOM exists).                        | Slightly higher (Virtual DOM + Real DOM).         |
| Used By      | Vanilla JavaScript, traditional web pages.             | React and similar libraries/frameworks.           |


Real DOM: Rewrite the actual book page every time a word changes.

Virtual DOM: First edit a copy of the page, find the exact changed word, then update only that word in the real book.
