library

1. GSAP
2. Lenis

Framework

1. Nextjs
2. Angular
3. Remix


A **library** is a collection of pre-written functions or tools that you call whenever you need them, meaning **you control the flow of the program** and decide when to use the library's features. A **framework**, on the other hand, provides a predefined structure and controls the application's flow, meaning **the framework calls your code when needed**—a concept known as *Inversion of Control*. For example, **React** is primarily a library because you decide how and when to use its components, while frameworks like Angular or Next.js provide a complete application structure, routing, conventions, and lifecycle management that your code must follow.

# Q. RealDOM vs VirtualDOM
| Feature      | Real DOM                                               | Virtual DOM                                       |
| ------------ | ------------------------------------------------------ | ------------------------------------------------- |
| Definition   | Actual DOM tree maintained by the browser.             | Lightweight JavaScript representation of the DOM. |
| Update Speed | Slower because changes directly affect the browser UI. | Faster because changes are first made in memory.  |
| Re-rendering | Often updates large parts of the DOM.                  | Updates only the changed elements.                |
| Performance  | Can become slow for complex applications.              | More efficient for frequent UI updates.           |
| Memory Usage | Lower (only actual DOM exists).                        | Slightly higher (Virtual DOM + Real DOM).         |
| Used By      | Vanilla JavaScript, traditional web pages.             | React and similar libraries/frameworks.           |


Real DOM: Rewrite the actual book page every time a word changes.

Virtual DOM: First edit a copy of the page, find the exact changed word, then update only that word in the real book.

# Q. what is component based architecture?
| Component-Based Architecture            | Non-Component-Based Architecture            |
| --------------------------------------- | ------------------------------------------- |
| UI is divided into reusable components. | UI is written as one large page or script.  |
| Easy to reuse code.                     | Code duplication is common.                 |
| Easier to maintain.                     | Harder to maintain as project grows.        |
| Used by React, Vue, Angular.            | Common in traditional HTML/CSS/JS websites. |
| Better scalability.                     | Poor scalability.                           |


# Q. what is probs drilling?
The process of passing different data to same components.

Props Drilling is the process of passing data from a parent component to a deeply nested child component through multiple intermediate components that do not use the data themselves. It can make applications difficult to maintain and is commonly solved using React Context API or state management libraries like Redux or Zustand.

# Q. What is CSS framework?
A CSS framework is a pre-written collection of CSS files, utility classes, components, and responsive design tools that help developers build web applications faster without writing all the CSS from scratch.

Instead of writing:

.button {

  background-color: blue;
  
  color: white;
  
  padding: 10px 20px;
  
  border-radius: 8px;

}

You can simply use a framework class:

<button class="btn btn-primary">
    Login
</button>

# Q. What is Context API?
Context API is a built-in feature of React that allows you to share data between multiple components without passing props through every intermediate component.

It is mainly used to solve Props Drilling.

# Props vs Context API :-
| Props                           | Context API                           |
| ------------------------------- | ------------------------------------- |
| Data flows from parent to child | Data is shared through a Provider     |
| Suitable for a few levels       | Suitable for many components          |
| Can lead to props drilling      | Eliminates props drilling             |
| Simple for local communication  | Best for shared application-wide data |
