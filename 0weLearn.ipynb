#  Q. what is diff b/w library and framework ?
library

1. GSAP
2. Lenis
3. ReactJS

Framework

1. Nextjs
2. Angular
3. Remix


A **library** is a collection of pre-written functions or tools that you call whenever you need them, meaning **you control the flow of the program** and decide when to use the library's features. A **framework**, on the other hand, provides a predefined structure and controls the application's flow, meaning **the framework calls your code when needed**—a concept known as *Inversion of Control*. For example, **React** is primarily a library because you decide how and when to use its components, while frameworks like Angular or Next.js provide a complete application structure, routing, conventions, and lifecycle management that your code must follow.


# Q. how to import and export?
two types of exports
1. Named Export

You can export multiple variables, functions, or classes from a file, and they must be imported using the exact names.

2. Default Export

A file can have only one default export. When importing, you can give it any name.
